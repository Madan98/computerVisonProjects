```python


# Try loading the image
img = cv2.imread("test2.png")



```

```python
imageReader=easyocr.Reader(['en'],gpu=False)
```

```python
#detect text on image
text=imageReader.readtext(img)
print(text)
```

```python
minClar=0.25
for i in text:
    bbox,actualText,confidence=i
    print(i)
    if confidence>minClar:
        cv2.rectangle(img,bbox[0],bbox[2],(0,255,0),3)
        cv2.putText(img,actualText,bbox[0],cv2.FONT_HERSHEY_SIMPLEX,0.65,(255,0,0),2)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
```
